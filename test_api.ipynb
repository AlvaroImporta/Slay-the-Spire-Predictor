```python
import requests
import pandas as pd

api_url = "https://raw.githubusercontent.com/alexdriedger/SlayTheSpireFightPredictor/master/out/1593042045/data_1593042341.json"

response = requests.get(api_url)
data = response.json()
data
```

```python
len(data)
```

```python
character = []
cards = []
relics = []
ascension = []
floor = []


for i in range(len(data)):
    character.append(data[i]['character'])
    cards.append(data[i]['cards'])
    relics.append(data[i]['relics'])
    ascension.append(data[i]['ascension'])
    floor.append(data[i]['floor'])

dictionario = {

    'character': character,
    'cards': cards,
    'relics': relics,
    'ascension': ascension,
    'floor': floor
}

df = pd.DataFrame(dictionario)
```
