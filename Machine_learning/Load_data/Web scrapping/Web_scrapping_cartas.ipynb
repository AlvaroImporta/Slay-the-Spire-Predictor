```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urlparse
```

```python
def create_dict_images(soup_in):
    imagenes = {}

    for imagen in soup_in.find("tbody").find_all("tr"):
        for nombre in imagen.find_all("td"):
            if nombre.a is not None and nombre.a.text != "":
                nombre_var = nombre.a.text
            if nombre.find("figure")is not None:
                png_var = nombre.find("figure").a["href"]
                imagenes.update({nombre_var: png_var})
    return imagenes
```

```python
def download_img(images, character):
    carpeta_imagenes = f'imagenes_cartas/imagenes_{character}'
    if not os.path.exists(carpeta_imagenes):
        os.makedirs(carpeta_imagenes) 

    # Descargar cada imagen
    for nombre, url_imagen in images.items():
        # Obtener el contenido de la imagen
        respuesta_imagen = requests.get(url_imagen, stream=True)
        if respuesta_imagen.status_code == 200:
            # Crear la ruta completa para guardar la imagen
            ruta_imagen = os.path.join(carpeta_imagenes, nombre + '.jpg')
            # Guardar la imagen en el archivo
            with open(ruta_imagen, 'wb') as archivo_imagen:
                for chunk in respuesta_imagen.iter_content(1024):
                    archivo_imagen.write(chunk)
        else:
            print(f'No se pudo descargar la imagen {nombre}.')
    
    print("Proceso Terminado")
```

```python
def download_all_urls():
    urls = ["https://slay-the-spire.fandom.com/wiki/Ironclad_Cards",
            "https://slay-the-spire.fandom.com/wiki/Silent_Cards",
            "https://slay-the-spire.fandom.com/wiki/Defect_Cards",
            "https://slay-the-spire.fandom.com/wiki/Watcher_Cards",
            "https://slay-the-spire.fandom.com/wiki/Colorless_Cards",
            "https://slay-the-spire.fandom.com/wiki/Status",
            "https://slay-the-spire.fandom.com/wiki/Curse"]

    for url in urls:
        response = requests.get(url)
        print(bool(response))
        soup = BeautifulSoup(response.text, "html.parser")

        parsed_url = urlparse(url)
        # Obtener la Ãºltima parte del path
        last_part = parsed_url.path.split('/')[-1]

        dict_images = create_dict_images(soup)
        download_img(dict_images, last_part)
```

```python
def dict_all_urls():
    urls = ["https://slay-the-spire.fandom.com/wiki/Ironclad_Cards",
            "https://slay-the-spire.fandom.com/wiki/Silent_Cards",
            "https://slay-the-spire.fandom.com/wiki/Defect_Cards",
            "https://slay-the-spire.fandom.com/wiki/Watcher_Cards",
            "https://slay-the-spire.fandom.com/wiki/Colorless_Cards",
            "https://slay-the-spire.fandom.com/wiki/Status",
            "https://slay-the-spire.fandom.com/wiki/Curse"]
    
    dict_images = {}

    for url in urls:
        response = requests.get(url)
        print(bool(response))
        soup = BeautifulSoup(response.text, "html.parser")

        d = create_dict_images(soup)

        dict_images.update(d)

    return dict_images
```

```python
enlaces_img = dict_all_urls()
print(enlaces_img)
```
